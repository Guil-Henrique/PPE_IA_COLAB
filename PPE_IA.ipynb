```python
from ultralytics import YOLO
```

```python
data_yaml = """
path: /content/drive/MyDrive/YOLOv8_EPI_UNISALES_GUILHERME/Dataset

train: train/images
val: valid/images
test: test/images

names:
  0: boots
  1: gloves
  2: helmet
  3: human
  4: vest
"""

with open("data.yaml", "w") as f:
    f.write(data_yaml)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data="data.yaml",
    epochs=3,
    imgsz=640,
    batch=16,
    project="/content/drive/MyDrive/YOLOv8_EPI_UNISALES_GUILHERME",
    name="yolov8_epi_model"
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/YOLOv8_EPI_UNISALES_GUILHERME/yolov8_epi_model3/weights/best.pt")

#imagem de teste
img_path = "/content/drive/MyDrive/YOLOv8_EPI_UNISALES_GUILHERME/Dataset/test/images/KakaoTalk_20221018_001605157_12_jpg.rf.2b1c6387d375e7f1302c9997361f8a06.jpg"

results = model.predict(img_path, conf=0.25, save=True)
```
